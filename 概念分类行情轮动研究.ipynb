```python
import tushare as ts
stockCf =  ts.get_concept_classified()
stockCf.head()
```

```python
stockCf.drop_duplicates(subset='c_name')#云计算
```

```python
classes1 = stockCf[stockCf['c_name']=='云计算']
classes1
```

```python
volumeList = []
classes1['volume'] = range(62)
for index,row in classes1.iterrows():
#     print(index,row)
#     print(row['code'])
    temp = ts.get_k_data(row['code'],ktype = 'D',start='2018-03-09',end='2018-03-09')
    #row['volume']=temp.values[0][5]
    #print(temp.values[0][5])
    a = temp.values
    if len(a)==1:
        volumeList.append(a[0][5])
    else:
        volumeList.append(0)
print(volumeList)
#ts.get_k_data('300397',ktype = 'D',start='2018-03-09',end='2018-03-09')
#classes1['volume']=volumeList
classes1['volume'] = volumeList
classes1
```

```python
result = classes1.sort_values('volume',ascending=False)
result.to_csv('云计算成交量排行榜.csv')
```
